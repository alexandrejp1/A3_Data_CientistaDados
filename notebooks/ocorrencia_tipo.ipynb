# Problemas em voo estão relacionados com ocorrências com fatalidades? 
Ocorrências em Solo/Voo X Fatalidades.

Dos 3 tipos de ocorrências mais existentes, 81,8% está relacionado com problemas em voo.

```python
import pandas as pd
import numpy as np
from IPython.display import display
from PIL import Image
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ocorrencia_tipo=pd.read_csv('../dados/ocorrencia_tipo.csv', sep = ';')
aeronaves_filtradas=pd.read_csv('../dados/aeronave_filtrada.csv', sep = ',')
ocorrencias_filtradas=pd.read_csv('../dados/ocorrencias_filtradas.csv', sep = ',')
```

# Unindo dados das duas tabelas: tipo_ocorrências e ocorrências.

```python
tipo_ocorrencias = pd.merge(ocorrencia_tipo, ocorrencias_filtradas, left_on=['codigo_ocorrencia1'], right_on=['codigo_ocorrencia1'])
tipo_ocorrencias
```

```python
tipo_ocorrencias_aeronaves = pd.merge(tipo_ocorrencias, aeronaves_filtradas, left_on=['codigo_ocorrencia1'], right_on=['codigo_ocorrencia2'])
tipo_ocorrencias_aeronaves
```

# Ocorrências com fatalidades:
# Falha em motor tem muita participação nesses registros.

```python
filter_ocorrencias = tipo_ocorrencias_aeronaves['ocorrencia_tipo'].value_counts().head(4)
top4_ocorrencias_tipo = tipo_ocorrencias_aeronaves.query('ocorrencia_tipo in @filter_ocorrencias.index.tolist()')
top4_ocorrencias_tipo


values = top4_ocorrencias_tipo[top4_ocorrencias_tipo['aeronave_fatalidades_total'] > 1].ocorrencia_tipo.value_counts().values
labels = top4_ocorrencias_tipo[top4_ocorrencias_tipo['aeronave_fatalidades_total'] > 1].ocorrencia_tipo.value_counts().index
```

```python
fig1, ax1 = plt.subplots()

def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d})".format(pct, absolute)



_, _, autopcts = ax1.pie(values, labels=labels, autopct=lambda pct: func(pct, values), colors = ['#dd8452', '#4c72b0', '#55a868'])


plt.setp(autopcts, **{'color':'white', 'weight':'bold', 'fontsize':12})


ax1.axis('equal')

plt.show()
```
