# O fator humano é responsável pela maioria das ocorrências?
Fatores operacionais são responsáveis por 63.8% das ocorrências, enquanto fatores humanos 33.2%.

# É possível dizer que o fator humano está relacionado com as fatalidades das ocorrências?
Sim, fatores humanos estão presentes em maior quantidade nas ocorrências com fatalidades.

```python
import pandas as pd
import numpy as np
from IPython.display import display
from PIL import Image
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fator_contribuinte=pd.read_csv('../dados/fator_contribuinte.csv', sep = ';')
ocorrencias_filtradas=pd.read_csv('../dados/ocorrencias_filtradas.csv', sep = ',')
aeronaves_filtradas=pd.read_csv('../dados/aeronave_filtrada.csv', sep = ',')
fator_contribuinte
```

# Unindo dados das duas tabelas: fator_contribuinte e ocorrências.

```python
fatorcontribuinte_ocorrencias = pd.merge(fator_contribuinte, ocorrencias_filtradas, left_on=['codigo_ocorrencia3'], right_on=['codigo_ocorrencia1'])
fatorcontribuinte_ocorrencias
```

# Área de Investigação

```python
fatorcontribuinte_ocorrencias.fator_area.value_counts()
```

```python
fator_contribuinte_graph = fatorcontribuinte_ocorrencias.groupby(['fator_area', 'ocorrencia_classificacao']).size().reset_index().pivot(columns='ocorrencia_classificacao', index='fator_area', values=0)
fator_contribuinte_graph = fator_contribuinte_graph.fillna(0)
fator_contribuinte_graph
```

## Panorama das ocorrências por Área de Investigação


```python
plt.figure(figsize = (15,8))

rc={'font.size': 12, 'axes.labelsize': 24, 'legend.fontsize': 40, 
    'axes.titlesize': 24, 'xtick.labelsize': 12, 'ytick.labelsize': 12}

sns.set(rc=rc)

ax = fator_contribuinte_graph.plot(kind='bar', stacked=True, figsize=(15, 8))

plt.legend(fontsize='x-large')
plt.xlabel("Área de Investigação")
plt.ylabel("Contagem Classificação")
plt.title("Classificação das ocorrências por Área de Investigação")
plt.xticks(rotation=45)

bottom = np.zeros(len(fator_contribuinte_graph))

for i, col in enumerate(fator_contribuinte_graph.columns):
    ax.bar(fator_contribuinte_graph.index, fator_contribuinte_graph[col], bottom=bottom, label=col)
    bottom += np.array(fator_contribuinte_graph[col])

# Somando as linhas dos dados para obter o valor total de cada barra.
totals = fator_contribuinte_graph.sum(axis=1)

# Add labels to each bar.
for i, total in enumerate(totals):
    ax.text(totals.index[i], total + 1, round(total), ha='center',
          weight='bold')

plt.show()
```

Para melhor análise: fiz a união das classes minoritárias.

Juntei as categorias de *, Fator Material e Outro em uma única (outros)

```python
labels = fatorcontribuinte_ocorrencias.fator_area.value_counts().index.tolist()
labels = [labels[0], labels[1], 'OUTRO']
labels
```

```python
values = fatorcontribuinte_ocorrencias.fator_area.value_counts().values
values = [values[0], values[1], sum(values[-3:])]
values
```

## Panorama das ocorrências por Área de Investigação

```python
fig1, ax1 = plt.subplots()

def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d})".format(pct, absolute)


_, _, autopcts = ax1.pie(values, labels=labels, autopct=lambda pct: func(pct, values), colors = ['#dd8452', '#4c72b0', '#55a868'])


plt.setp(autopcts, **{'color':'white', 'weight':'bold', 'fontsize':12})


ax1.axis('equal')

plt.show()
```
